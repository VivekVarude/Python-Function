```python
#1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
#the list.

def Sq_num(num):
    num1 = 0
    for val in num:
        num1 = val**2 + num1
    return num1
list2 = [1,2,3,4,5,6,7]
result = Sq_num(list2)
print(result)
```

```python
#3. Implement a Python function that takes a list of integers and returns a new list containing the squares of
#each number.
def Sq_num(num):
    list_sq = []
    for val in num:
        num1 = val**2 
        list_sq.append(num1)
    return list_sq
list1 = [1,2,3,4,5,6,7]
result = Sq_num(list1)
print(result)
```

```python
#4. Write a Python function that checks if a given number is prime or not from 1 to 200.
def check_prime(num):
    if num<2:
        return False
    for i in range(2,int(num**0.5)+1):
        if num%i == 0:
            return False
    return True
for i in range(1,201):
    if check_prime(i):
        print(f"{i} is prime")
        
```

```python
#5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
#terms.
class Fibonacci:
    def __init__(self, n):
        self.n = n          # number of terms
        self.a = 0          # first term
        self.b = 1          # second term
        self.count = 0      # to track number of generated terms

    def __iter__(self):
        return self         # iterator object

    def __next__(self):
        if self.count >= self.n:
            raise StopIteration
        
        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            next_val = self.a + self.b
            self.a, self.b = self.b, next_val
            self.count += 1
            return next_val


# Using the iterator
fib = Fibonacci(10)     # generate 10 terms

for num in fib:
    print(num)
```

```python
#6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def power_of_two(n):
    for i in range(n + 1):
        yield 2 ** i

# Example: print powers of 2 up to exponent 8
for val in power_of_two(8):
    print(val)
```

```python
#8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple
data = [(3, 20), (1, 5), (4, 15), (2, 10)]

sorted_data = sorted(data, key=lambda x: x[1])

print(sorted_data)
```
